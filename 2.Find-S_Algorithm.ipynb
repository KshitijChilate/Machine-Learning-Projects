### Aim: To implement the Find-S algorithm to find a maximally specific hypothesis on a given dataset.

### Steps Involved In Find-S :

#### 1. Start with the most specific hypothesis.  
#### h = {ϕ, ϕ, ϕ, ϕ, ϕ, ϕ}
#### 2. Take the next example and if it is negative, then no changes occur to thehypothesis.
#### 3. If the example is positive and we find that our initial hypothesis is too specific then we update our current hypothesis to a general condition.
#### 4. Keep repeating the above steps till all the training examples are complete.
#### 5. After we have completed all the training examples we will have the final hypothesis when can use to classify the new examples.

```python
feature=np.array(df)[:,:-1]
feature
```

```python
target=np.array(df)[:,-1]
target
```

```python
def find_S(df):
    hypothesis = ['0'] * feature.shape[1]      #most specific {all values phi}
                                               # here feature.shape[1] is giving value of 6 columns leaving the last target one

    for i in range(len(feature)):
        if target[i] == 1:                         #all +ve values hypothesis accepted

          for j in range(len(hypothesis)):
                if hypothesis[j] == '0':           #all -ve values hypothesis rejected
                    hypothesis[j] = feature[i][j]  #h2=h3
                elif hypothesis[j] != feature[i][j]: #replacing with '?'
                    hypothesis[j] = '?'
    return hypothesis

print(find_S(df))
```
